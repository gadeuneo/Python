```python
'''
For profile.json GroupMe files.
    Takes in dictionary and outputs key
    values. Also splits purchases dictionary
    and output list key values.
''' 

def pro(d):
    k = d.keys()
    for j in k:
        print(str(j)+" : "+str(d[j]))




'''
For phone_number_changes.json GroupMe files.
    Takes in list of dictionaries and outputs key
    values.
''' 


def phone(l):
    for d in l:
        k = d.keys()
        for j in k:
            print(str(j)+" : "+str(d[j]))
        print("")



'''
For powerup.json GroupMe files.
    Takes in dictionary and outputs key
    values. Also splits purchases dictionary
    and output list key values.
''' 


def power(d):
    k = d.keys()
    for j in k:
        if (str(j)=="purchases"):
            print(str(j)+":")
            l = d[j]
            for item in l:
                print("\t"+str(item))
        else:
            print(str(j)+" : "+str(d[j]))



'''
For mixpanel_user_profile.json GroupMe files.
    Takes in dictionary and outputs key
    values.
''' 


def mix(d):
    k = d.keys()
    for j in k:
        print(str(j)+" : "+str(d[j]))




'''
For push_registration.json GroupMe files.
    Takes in list of dictionaries and outputs key
    values.
''' 


def push(l):
    for d in l:
        k = d.keys()
        for j in k:
            print(str(j)+" : "+str(d[j]))
        print("")



'''
For install.json GroupMe files.
    Takes in list of dictionaries and outputs key
    values.
''' 


def install(l):
    for d in l:
        k = d.keys()
        for j in k:
            print(str(j)+" : "+str(d[j]))
        print("")



'''
For device.json GroupMe files.
    Takes in list of dictionaries and outputs key
    values.
''' 


def dev(l):
    for d in l:
        k = d.keys()
        for j in k:
            print(str(j)+" : "+str(d[j]))
        print("")



'''
For mine.json GroupMe files.
    Takes in dictionary and outputs key
    values. Also breaks down response dictionary
    key values.
''' 


def mine(d):
    k = d.keys()
    for j in k:
        if (str(j)=="response"):
            print(str(j)+":")
            res = d[j]
            for r in res.keys():
                print("\t"+str(r)+" : "+str(res[r]))
        else:
            print(str(j)+" : "+str(d[j]))



'''
For for_me.json GroupMe files.
    Takes in dictionary and outputs key
    values. Also breaks down response dictionary
    key values.
''' 

def forMe(d):
    k = d.keys()
    for j in k:
        if (str(j)=="response"):
            print(str(j)+":")
            res = d[j]
            for r in res.keys():
                print("\t"+str(r)+" : "+str(res[r]))
        else:
            print(str(j)+" : "+str(d[j]))



'''
For everyone.json GroupMe files.
    Takes in dictionary and outputs key
    values. Also breaks down response dictionary
    key values.
'''    
    
def every(d):
    k = d.keys()
    for j in k:
        if (str(j)=="response"):
            print(str(j)+":")
            res = d[j]
            for r in res.keys():
                print("\t"+str(r)+" : "+str(res[r]))
        else:
            print(str(j)+" : "+str(d[j]))




'''
For message.json GroupMe files.
    Takes in json file as input (l), which is a list
    of dictionaries (d). Keys for d (k) are split into
    their values (j). For the event key (eve), it is
    further split into more dictionaries (e). Then there
    are more nested dictionaries data (dat) that contain
    more dictionaries for 'adder_user' (a) and 'added_users'
    which contains another list of dictionaries (lst) and
    has values (l) which contain a final dictionary (lstk)
    and their values (lk).
'''


def mess(l):
    for d in l:
        k = d.keys()
        for j in k:
            if (str(j)=="created_at"):
                print(str(j)+" : "+str(dt.datetime.utcfromtimestamp(d[j])))
            elif (str(j)=="event"):
                print(str(j)+":")
                eve = d[j]
                for e in eve.keys():
                    if (str(e)=="data"):
                        print("\t"+str(e)+":")
                        dat = eve[e]
                        for a in dat.keys():
                            if (a=="adder_user"):
                                print("\t"+"\t"+str(a))
                                use = dat[a]
                                for u in use.keys():
                                    print("\t"+"\t"+"\t"+str(u)+" : "+str(use[u]))
                            if (a=="added_users"):
                                print("\t"+"\t"+str(a)+":")
                                lst = dat[a]
                                for l in lst:
                                    lstk = l
                                    for lk in lstk.keys():
                                        print("\t"+"\t"+"\t"+str(lk)+" : "+str(lstk[lk]))
                    else:
                        print("\t"+str(e)+" : "+str(eve[e]))
            else:
                print(str(j)+" : "+str(d[j]))
        print("")


"""
For conversation.json files from GroupMe:
    Takes in json file as input, converted to dict (d).
    For all the keys (k) in (d), we print out the value (j)
and any data associated with it converted to human readable format.
"""

def convo(d):
    k = d.keys()
    for j in k:
        if (str(j)=="created_at"):
            print(str(j)+" : "+str(dt.datetime.utcfromtimestamp(d[j])))
        elif (str(j)=="updated_at"):
            print(str(j)+" : "+str(dt.datetime.utcfromtimestamp(d[j])))
        elif (str(j)=="messages"):
            print(str(j)+":")
            mes = d[j]
            for l in mes:
                if(str(l)=="last_message_created_at"):
                    print("\t"+str(l)+" : "+str(dt.datetime.utcfromtimestamp(mes[l])))
                elif(str(l)=="preview"):
                    print("\t"+str(l)+":")
                    pre = mes[l]
                    for p in pre:
                        print("\t"+"\t"+str(p)+" : "+str(pre[p]))
                else:
                    print("\t"+str(l)+" : "+str(mes[l]))
        elif (str(j)=="members"):
            print(str(j)+":")
            for i in range(len(d[j])):
                m = d[j][i]
                for i in m:
                    print("\t"+str(i)+" : "+str(m[i]))
                print("")
        else:
            print(str(j)+" : "+str(d[j]))
```

```python
# Master Break Down
def decode(files):
    for i in range(len(files)):
        with open(os.path.join(files[i]),encoding="utf8") as f:
            d = json.load(f)
            if "conversation.json" in files[i]:
                print(files[i])
                convo(d)
            elif "message.json" in files[i]:
                print(files[i])
                mess(d)
            elif "everyone.json" in files[i]:
                print(files[i])
                every(d)
            elif "for_me.json" in files[i]:
                print(files[i])
                forMe(d)
            elif "mine.json" in files[i]:
                print(files[i])
                mine(d)
            elif "device.json" in files[i]:
                print(files[i])
                dev(d)
            elif "installation.json" in files[i]:
                print(files[i])
                install(d)
            elif "push_registration.json" in files[i]:
                print(files[i])
                push(d)
            elif "mixpanel_user_profile.json" in files[i]:
                print(files[i])
                mix(d)
            elif "powerup.json" in files[i]:
                print(files[i])
                power(d)
            elif "phone_number_changes.json" in files[i]:
                print(files[i])
                phone(d)
            elif "profile.json" in files[i]:
                print(files[i])
                pro(d)
                
    print("Decoding Done")
    sys.exit()
```
